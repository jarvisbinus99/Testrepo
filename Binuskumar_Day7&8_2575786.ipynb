# 1. Write a Python program to find a target values in a list using linear search with following steps
a. Initialize the list to store the input elements.

b. Initialize found=False.

C Enter the item to be searched (match_item).

d. For each element in the list

```python
data = []
n_str = input('Enter no of elements to be inserted separated by spaces: ')
n_list = n_str.split()


n = int(n_list[0])

for i in range(0, n):
    x = int(input(f'Enter the Element {i + 1}: '))
    data.append(x)

e = int(input('Enter the value to be searched: '))
pos = 0
found = False

while pos < n and not found:
    if data[pos] == e:
        found = True
    else:
        pos += 1

if found:
    print('Element %d Found at Position %d' % (e, pos + 1))
else:
    print('Element %d not found in the list' % e)

print(data)
```

# 2. Write a Python program to implement binary search to find the target values from the list:
a. Create a separate function to do binary search. b. Get the number of inputs from the user

C. Store the inputs individually in a list.

d. In binary search function at first sort the list in order to start the search from middle of the list.

e Compare the middle element to right and left elements to search target element

f. If greater, move to right of list or else move to another side of the list.

g. Print the result along with the position of the element.

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        mid_value = arr[mid]
        
        if mid_value == target:
            return f"Element {target} found at position {mid}"
        elif mid_value < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return f"Element {target} not found in the list"

input_list = [int(x) for x in input("Enter a list of numbers separated by spaces: ").split()]
target_value = int(input("Enter the target value to search for: "))
result = binary_search(input_list, target_value)
print(result)
```

# 3. Write a Python program for sorting a list of elements using selection sort algorithm:
a. Assume two lists: Sorted list- Initially empty and Unsorted List-Given input list.

b. In the first iteration, find the smallest element in the unsorted list and place it in the sortedlist. c.In the second iteration, find the smallest element in the unsorted list and place it in the

correct position by comparing with the element in the sorted list. d. In the third iteration, again find the smallest element in the unsorted list and place it in the

correct position by comparing with the elements in the sorted list. e. This process continues till the unsorted list becomes empty. f Display the sorted elements

```python
def selection_sort(arr):
    sorted_list = []  # Initialize an empty sorted list

    while arr:  # Continue until the unsorted list becomes empty
        min_index = 0

        # Find the index of the minimum element in the unsorted list
        for i in range(1, len(arr)):
            if arr[i] < arr[min_index]:
                min_index = i

        # Remove the minimum element from the unsorted list and append it to the sorted list
        sorted_list.append(arr.pop(min_index))

    return sorted_list

# Test the selection_sort function
input_list = [12, 4, 78, 35, 19, 7, 56]
sorted_list = selection_sort(input_list)
print("Original List:", input_list)
print("Sorted List:", sorted_list)
```

# Q5.Write a Python program that performs merge sort on a list of numbers:
a. Divide: If the given array has zero or one element, return.

Otherwise
ii. Divide the input list in to two halves each containing half of the elements. i.e. left half and right half.

b. Conquer: Recursively sort the two lists (left half and right half).

a. Call the merge sort on left half.

b. Call the merge sort on right half.

C. Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequence.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Divide the input list into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursively sort the two halves
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # Merge the sorted halves
    merged = []
    left_idx, right_idx = 0, 0

    while left_idx < len(left_half) and right_idx < len(right_half):
        if left_half[left_idx] < right_half[right_idx]:
            merged.append(left_half[left_idx])
            left_idx += 1
        else:
            merged.append(right_half[right_idx])
            right_idx += 1

    # Append remaining elements (if any)
    merged.extend(left_half[left_idx:])
    merged.extend(right_half[right_idx:])

    return merged

def get_user_input():
    try:
        input_list = input("Enter a list of numbers separated by spaces: ")
        input_list = [int(x) for x in input_list.split()]
        return input_list
    except ValueError:
        print("Invalid input. Please enter a list of numbers.")
        return get_user_input()

if __name__ == "__main__":
    print("Welcome to Merge Sort!")
    
    input_list = get_user_input()
    sorted_list = merge_sort(input_list)
    
    print("Original array:", input_list)
    print("Sorted array:", sorted_list)
```

# 6. Write a Python script to perform the following operations on a singly linked list
a. Create a list

b. Find the smallest element from the list

C.Insert an element if it is not a duplicate element

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    def create_list(self, elements):
        for element in elements:
            new_node = Node(element)
            if not self.head:
                self.head = new_node
            else:
                current = self.head
                while current.next:
                    current = current.next
                current.next = new_node

    def find_smallest(self):
        if not self.head:
            return None

        smallest = self.head.data
        current = self.head.next

        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next

        return smallest

    def insert_if_not_duplicate(self, element):
        if not self.head:
            self.head = Node(element)
            return

        current = self.head
        while current.next:
            if current.data == element:
                return  
            current = current.next

        if current.data != element:
            current.next = Node(element)

    def display_reverse(self):
        if not self.head:
            return []

        elements = []
        current = self.head
        while current:
            elements.append(current.data)
            current = current.next

        return elements[::-1]

if __name__ == "__main__":
    elements = [int(x) for x in input("Enter elements separated by spaces: ").split()]
    linked_list = LinkedList()
    linked_list.create_list(elements)

    smallest = linked_list.find_smallest()
    if smallest is not None:
        print(f"The smallest element in the list is: {smallest}")
    else:
        print("The list is empty.")

    element_to_insert = int(input("Enter an element to insert: "))
    linked_list.insert_if_not_duplicate(element_to_insert)

    reversed_elements = linked_list.display_reverse()
    if reversed_elements:
        print("Elements in reverse order:", reversed_elements)
    else:
        print("The list is empty.")
```

# 7.Write a python program to implement the various operations for Stack ADT 1.) Push ii.) Pop iii.) Display.

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stack = [None] * capacity
        self.top = -1  # Initialize the top pointer to -1

    def is_full(self):
        return self.top == self.capacity - 1

    def is_empty(self):
        return self.top == -1

    def push(self, item):
        if self.is_full():
            print("Stack is full. Cannot push the element.")
        else:
            self.top += 1
            self.stack[self.top] = item

    def pop(self):
        if self.is_empty():
            print("Stack is empty. Cannot pop an element.")
            return None
        else:
            item = self.stack[self.top]
            self.top -= 1
            return item

    def peek(self):
        if self.is_empty():
            print("Stack is empty. Cannot peek.")
            return None
        else:
            return self.stack[self.top]

# Example usage:
stack = Stack(7)
stack.push(1)
stack.push(2)
stack.push(3)
print("Top element:", stack.peek())  # Output: 3
stack.push(4)
stack.push(5)
```

# 9 Write a program in python to convert the following infix expression to its postfix form using push and pop operations of a Stack
a. A/B^C+DE-FG

b. (B^2-4AC)^(1/2) (100)

```python

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0


def infix_to_postfix(expression):
    precedence = {'+':1, '-':1, '*':2, '/':2, '^':3}
    postfix = []
    operator_stack = Stack()

    for char in expression:
        if char.isalnum():  # Operand, add to the postfix expression
            postfix.append(char)
        elif char == '(':  # Left parenthesis, push to stack
            operator_stack.push(char)
        elif char == ')':  # Right parenthesis, pop and append operators until a matching '(' is found
            while not operator_stack.is_empty() and operator_stack.peek() != '(':
                postfix.append(operator_stack.pop())
            operator_stack.pop()  # Pop the '('
        elif char in precedence:  # Operator
            while (not operator_stack.is_empty() and operator_stack.peek() != '(' and
                   precedence[char] <= precedence.get(operator_stack.peek(), 0)):
                postfix.append(operator_stack.pop())
            operator_stack.push(char)

    # Pop any remaining operators from the stack
    while not operator_stack.is_empty():
        postfix.append(operator_stack.pop())

    return ''.join(postfix)

# Test cases
if __name__ == "__main__":
    # a. Infix expression: A/B^C+D*E-F*G
    infix_a = "A/B^C+D*E-F*G"
    postfix_a = infix_to_postfix(infix_a)
    print("Postfix expression for a:", postfix_a)

    # b. Infix expression: (B^2-4*A*C)^(1/2) (100)
    infix_b = "(B^2-4*A*C)^(1/2) (100)"
    postfix_b = infix_to_postfix(infix_b)
    print("Postfix expression for b:", postfix_b)
```
